# RAGAS EVALUATION

```python
import ragas
from datasets import Dataset
from langchain_openai import ChatOpenAI, OpenAIEmbeddings

print("RAGAS installed successfully")
```

```python
import json
import os
import pandas as pd
from datasets import Dataset
import dotenv

dotenv.load_dotenv()

from ragas import evaluate
from ragas.metrics import (
    faithfulness,
    answer_correctness,
    context_precision,
    context_recall
)

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_KEY")

with open("rag_evaluation_scored_results.json", "r", encoding="utf-8") as f:
    rag_eval_results = json.load(f)
    print(f"Loaded {len(rag_eval_results)} evaluation results")
```

```python
ragas_data = {
    "question": [],
    "answer": [],
    "contexts": [],
    "ground_truth": []
}

for item in rag_eval_results:
    ragas_data["question"].append(item["question"])
    ragas_data["answer"].append(item["generated_answer"])
    ragas_data["contexts"].append(item["retrieved_contexts"])
    ragas_data["ground_truth"].append(item["expected_answer"])

ragas_dataset = Dataset.from_dict(ragas_data)

print(ragas_dataset)
```

```python
small_dataset = ragas_dataset.select(range(10))

test_scores = evaluate(
    small_dataset,
    metrics=[
        faithfulness,
        answer_correctness,
        context_precision,
        context_recall
    ]
)

test_df = test_scores.to_pandas()
display(test_df)
```

```python
ragas_scores = evaluate(
    ragas_dataset,
    metrics=[
        faithfulness,
        answer_correctness,
        context_precision,
        context_recall
    ]
)

ragas_df = ragas_scores.to_pandas()
display(ragas_df)
```

```python
print("RAGAS Evaluation Summary")
print("=" * 50)

ragas_metric_cols = [
    "faithfulness",
    "answer_correctness",
    "context_precision",
    "context_recall"
]

for col in ragas_metric_cols:
    print(f"{col}: {ragas_df[col].mean():.2%}")
```

```python
ragas_df.to_csv("ragas_evaluation_results.csv", index=False)
print("Saved ragas_evaluation_results.csv")
```

RAGAS evaluation was also conducted on the 75 validated question-answer pairs. The system achieved 90.09% faithfulness, 59.46% answer correctness, 59.05% context precision, and 68.93% context recall. The high faithfulness score indicates that the generated answers were generally well-grounded in the retrieved context. However, the lower answer correctness and context precision scores suggest that some retrieved chunks were only partially relevant, which affected how closely the generated answers matched the manually validated expected answers.
